```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
# correlation with heatmap
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

```python
#joinplot
#univariate analysis
#a join plot allows to study the relationship between 2 numeric variables.
#the central chart display their correlation it is usually a sctterplot a hextbin plot ,a 2d histogram or a 2d density plot
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#pairplot
#scatterplot #more then two independent features
#in which one variable in the same data is matched with another variables value,like this pair plots are just elaborations on this
#showing all variables paired with all the other variables
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex') #plot the grape base on sex column
```

```python
df['smoker'].value_counts()
```

```python
#distplot
#check the distribution of the columns feature

sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=True,bins=10)
```

```python
sns.displot(df['tip'],kde=False,bins=10)#kde karnel density
```

```python
#categorical plots
#boxplot
#violinplot
#countplot
#barplot
```

```python
#count

sns.countplot(x='sex',data=df)#x axis
```

```python
sns.countplot(y='day',data=df)#y axis
```

```python
#barplot

#sns.barplot(x='total_bill',y='smoker' ,data=df)
sns.barplot(y='total_bill',x='smoker' ,data=df)
```

```python
#boxplot
#presents information from a five-number summary

sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
#sns.boxplot(data=df,orient='v')
sns.boxplot(data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
#violinplot
#violin plot helps us to see both the distribution of data interms of kernel density estimation and the box plot

sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
