```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

a=np.arange(40,50)
b=np.arange(50,60)
```

```python
#scatter

plt.scatter(a,b,c='r')
plt.title('graph in 2d')
plt.xlabel('x axis')
plt.ylabel('y axis')

#save the image

plt.savefig('plot1.jpg')
```

```python
# plot

b=a*a
#plt.plot(a,b,'go')#bo blue #go green
#plt.plot(a,b,'r*-')


plt.plot(a,b,'r*',linestyle='dashed',linewidth=2,markersize=12)#dashaed ro--

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('plot2')
```

```python
#creating sub plots
#press shift +tab  to get about function
#*args : int, (int, int, *index*), or .SubplotSpec, default: (1, 1, 1)
    #The position of the subplot described by one of Three integers (*nrows*, *ncols*, *index*)

plt.subplot(4,3,1)
plt.plot(a,b,'bo')

plt.subplot(3,2,2)
plt.plot(a,b,'go')
```

```python
plt.subplot(3,2,2)
plt.plot(a,b,'ro')

plt.subplot(2,3,1)  #index means place
plt.plot(a,b,'go')
```

```python
x=np.array([9,7,4])
y=3**x+5

plt.title("matplotlib demo")
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.plot(x,y)

plt.show()
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave')

plt.plot(x,y)
plt.show()
```

```python
k=np.arange(0,5*np.pi,0.1)

x=np.sin(k)
y=np.cos(k)

plt.plot(x,y)
```

```python
#subplots

k=np.arange(0,5*np.pi,0.1)

x=np.sin(k)
y=np.cos(k)


plt.subplot(2,1,2)
plt.plot(k,x,'r--')
plt.title('sine')
```

```python
plt.subplot(3,2,4)
plt.plot(k,y,'g--')
plt.title('cosine')
```

```python
#bar plot

x=[3,4,6]
y=[6,7,4]

x1=[8,7,6]
y1=[9,7,3]

plt.bar(x,y,color='r')
plt.bar(x1,y1,color='g')

plt.title('bar graph')

plt.xlabel('x axis')
plt.ylabel('y axis')

```

```python
#histograms

a=np.array([12,24,31,42,55,37,69,78,32,93])
plt.hist(a,color='r')#all div by 20 bins=20
plt.title("histogram")
plt.show()
```

```python
#box plot help to find %

data=[np.random.normal(0,std, 100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
#box plot help to find %

data=[np.random.normal(0,std, 100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False)#colour
```

```python
#pie chart

labels='python','c++','ruby','java'
size=[67,98,55,34]# percentage
colors=['gold','pink','red','green']
explode=(0.6,0,0,0)#explode 1st slice

#plot

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
